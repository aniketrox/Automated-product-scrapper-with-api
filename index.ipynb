```python
import re

def extract_numbers_from_url(url):
    # Define a regex pattern to match digits at the end of the URL
    pattern = r'\d+$'
    
    # Use re.search to find the pattern in the URL
    match = re.search(pattern, url)
    
    # If a match is found, extract the digits and return them
    if match:
        numbers = match.group()
        return str(numbers)
    else:
        return None  # Return None if no numbers found

# Example usage:
url = "https://themeforest.net/item/megaone-one-page-parallax/25651324"
numbers = extract_numbers_from_url(url)
if numbers is not None:
    print("Numbers at the end of the URL:", numbers)
else:
    print("No numbers found at the end of the URL.")
```

```python
def download_image(image_url, save_folder, file_name):
    if not os.path.exists(save_folder):
        os.makedirs(save_folder)
    file_path = os.path.join(save_folder, file_name)
    
    response = requests.get(image_url)

    if response.status_code == 200:
        with open(file_path, 'wb') as f:
            f.write(response.content)
```
